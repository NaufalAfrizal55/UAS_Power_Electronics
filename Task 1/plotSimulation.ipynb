```python
import matplotlib.pyplot as plt
import csv
```

```python
# Ganti dengan nama file CSV kamu
filename = 'Vout_OpenLoop_BuckConverter.csv'

# List untuk menyimpan data
time = []
voltage = []

# Membaca data CSV
with open(filename, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if len(row) >= 2:  # Pastikan ada dua kolom
            try:
                t = float(row[0])
                v = float(row[1])
                time.append(t)
                voltage.append(v)
            except ValueError:
                continue  # Lewati baris jika tidak bisa dikonversi

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(time, voltage, label='Tegangan', color='blue')
plt.title('Plot Tegangan terhadap Waktu')
plt.xlabel('Waktu (s)')
plt.ylabel('Tegangan (V)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
